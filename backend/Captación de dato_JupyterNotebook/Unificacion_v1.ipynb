## Limpieza Transfermarkt

```python
import pandas as pd

# Cargar el archivo de Transfermarkt
transfermarkt_path = "/Users/macmontxinho/Desktop/Master Python/TFM/Transfermark.xlsx"
df_transfermarkt = pd.read_excel(transfermarkt_path)

# Paso 1: Eliminar el texto entre paréntesis en la columna 'Fecha Nac' y crear la nueva columna 'Nacimiento'
df_transfermarkt['Nacimiento'] = df_transfermarkt['Fecha Nac'].str.replace(r'\(.*\)', '', regex=True).str.strip()

# Paso 2: Crear la columna 'Año' extrayendo solo el año de la columna 'Nacimiento'
df_transfermarkt['Año'] = df_transfermarkt['Nacimiento'].str.split('/').str[-1]

# Paso 3: Eliminar la columna original 'Fecha Nac'
df_transfermarkt = df_transfermarkt.drop(columns=['Fecha Nac'])

# Verificar los cambios
print("Primeras filas después de los cambios:")
display(df_transfermarkt[['Nacimiento', 'Año']].head())

# Verificar las columnas
print("Columnas en Transfermarkt después de los cambios:", df_transfermarkt.columns.tolist())
```

```python
# La columna 'Edad' ya tiene valores numéricos pero puede tener decimales, por lo que la convertimos a entero
df_transfermarkt['Edad'] = df_transfermarkt['Edad'].astype(int)

# Convertir la columna 'Nacimiento' a formato datetime, asegurándonos de que está correctamente interpretada
df_transfermarkt['Nacimiento'] = pd.to_datetime(df_transfermarkt['Nacimiento'], format='%d/%m/%Y', errors='coerce')

# Asegurarnos de que solo los valores numéricos se mantengan en la columna 'Año'
df_transfermarkt['Año'] = pd.to_numeric(df_transfermarkt['Año'], errors='coerce')

# Verificar los valores que no se pudieron convertir a entero
print("Valores no válidos en 'Año':")
print(df_transfermarkt[df_transfermarkt['Año'].isna()])

# Ahora que los valores no válidos han sido manejados, podemos convertir a entero
df_transfermarkt['Año'] = df_transfermarkt['Año'].fillna(0).astype(int)

# Verificar que la columna 'Año' ha sido convertida correctamente
print("\nDatos después de la conversión:")
print(df_transfermarkt[['Año']].head())
```

#### Revisando el archvio de Fbref, vimos que las columnas con % solo había valores 0 y 0.01, por lo que las identificamos y calculamos aquellas donde era posible (Todas excepto %Tkld_exito)

```python
import pandas as pd

# Filtrar las columnas que contienen '%'
columnas_con_porcentaje = [col for col in df_fbref.columns if '%' in col]

# Crear las nuevas columnas con el sufijo '_calculado' para cada porcentaje
for col in columnas_con_porcentaje:
    if col == '%Disparos_arco':
        # Calcular %Disparos_arco = Disparos_arco / Disparos * 100
        df_fbref[f'{col}_calculado'] = ((df_fbref['Disparos_arco'] / df_fbref['Disparos']) * 100).fillna(0).round(2)
    elif col == '%Pases_exito':
        # Calcular %Pases_exito = Pases_exito / Pases_intentados * 100
        df_fbref[f'{col}_calculado'] = ((df_fbref['Pases_exito'] / df_fbref['Pases_intentados']) * 100).fillna(0).round(2)
    elif col == '%Pase_corto_exito':
        # Calcular %Pase_corto_exito = Pase_corto_exito / Pase_corto_intentado * 100
        df_fbref[f'{col}_calculado'] = ((df_fbref['Pase_corto_exito'] / df_fbref['Pase_corto_intentado']) * 100).fillna(0).round(2)
    elif col == '%Pase_medio_exito':
        # Calcular %Pase_medio_exito = Pase_medio_exito / Pase_medio_intentado * 100
        df_fbref[f'{col}_calculado'] = ((df_fbref['Pase_medio_exito'] / df_fbref['Pase_medio_intentado']) * 100).fillna(0).round(2)
    elif col == '%Pase_largo_exito':
        # Calcular %Pase_largo_exito = Pase_largo_exito / Pase_largo_intentado * 100
        df_fbref[f'{col}_calculado'] = ((df_fbref['Pase_largo_exito'] / df_fbref['Pase_largo_intentado']) * 100).fillna(0).round(2)
    elif col == '%Desafios_ganados':
        # Calcular %Desafios_ganados = Desafios_ganados / Desafios_intentados * 100
        df_fbref[f'{col}_calculado'] = ((df_fbref['Desafios_ganados'] / df_fbref['Desafios_intentados']) * 100).fillna(0).round(2)
    elif col == '%Regates_exito':
        # Calcular %Regates_exito = Regates_exito / Regates_intentados * 100
        df_fbref[f'{col}_calculado'] = ((df_fbref['Regates_exito'] / df_fbref['Regates_intentados']) * 100).fillna(0).round(2)
    elif col == '%Aereo_ganado':
        # Calcular %Aereo_ganado = Arereo_ganado / (Arereo_ganado + Aereo_perdido) * 100
        df_fbref[f'{col}_calculado'] = ((df_fbref['Arereo_ganado'] / (df_fbref['Arereo_ganado'] + df_fbref['Aereo_perdido'])) * 100).fillna(0).round(2)
    elif col == '%Tkld_exito':
        # No podemos calcularlo porque falta la columna de entradas exitosas
        df_fbref[f'{col}_calculado'] = 0  # o cualquier otro valor que tenga sentido en tu análisis

# Mostrar el DataFrame con las columnas ordenadas y las primeras filas
print("\nPrimeras filas con las columnas calculadas:")
display(df_fbref.head())  # Usar display en Jupyter
```

### Realizamos la estrategia de coincidencias de múltiples pasos, primero buscando coincidencias por equipo y año de nacimiento, y luego aplicando RapidFuzz solo a aquellos jugadores que no coincidan en esos campos

```python
from rapidfuzz import process
import pandas as pd

# 1. Realizar el join entre df_fbref y df_transfermarkt para obtener las columnas 'Valor Mercado' y 'Pos' (específica de Transfermarkt)
df_combinado = pd.merge(df_fbref, 
                        df_transfermarkt[['Jugador_Normalizado', 'Equipo_TM_Ajustado', 'Año', 'Valor Mercado', 'Pos']], 
                        how='left', 
                        left_on=['Jugador_Normalizado', 'Nacimiento', 'Squad'], 
                        right_on=['Jugador_Normalizado', 'Año', 'Equipo_TM_Ajustado'])

# Renombrar las columnas 'Pos_x' y 'Pos_y' para evitar confusión
df_combinado.rename(columns={'Pos_x': 'Pos_General_FBRef', 'Pos_y': 'Pos_Especifica_Transfermarkt'}, inplace=True)

# 2. Función para mapear el 'Valor Mercado' y la 'Pos_Especifica' usando equipo y año de nacimiento
def mapear_por_equipo_y_año(row, df_transfermarkt):
    equipo_fbref = row['Squad']
    año_fbref = row['Nacimiento']
    
    # Buscar coincidencias en Transfermarkt por equipo y año de nacimiento
    coincidencias = df_transfermarkt[(df_transfermarkt['Equipo_TM_Ajustado'] == equipo_fbref) & 
                                     (df_transfermarkt['Año'] == año_fbref)]
    
    # Si encontramos una coincidencia exacta, obtenemos el 'Valor Mercado' y 'Pos_Especifica'
    if not coincidencias.empty:
        # Asumimos que solo hay una coincidencia
        valor_mercado = coincidencias['Valor Mercado'].values[0]
        pos_especifica = coincidencias['Pos'].values[0]  # Usamos 'Pos' directamente
        return valor_mercado, pos_especifica
    return None, None

# 3. Aplicar la función para filtrar por equipo y año de nacimiento
df_combinado[['Valor Mercado', 'Pos_Especifica_Transfermarkt']] = df_combinado.apply(
    lambda row: pd.Series(mapear_por_equipo_y_año(row, df_transfermarkt)) 
    if pd.isna(row['Valor Mercado']) else pd.Series([row['Valor Mercado'], row['Pos_Especifica_Transfermarkt']]),
    axis=1
)

# 4. Función para mapear el 'Valor Mercado' y 'Pos_Especifica' usando RapidFuzz (solo si no se ha asignado antes)
def mapear_por_nombre(row, df_transfermarkt, umbral_similitud=70):
    nombre = row['Jugador_Normalizado']
    
    # Buscar el jugador en df_transfermarkt con la mayor similitud en el nombre
    resultado = process.extractOne(nombre, df_transfermarkt['Jugador_Normalizado'], score_cutoff=umbral_similitud)
    
    # Si encontramos una coincidencia válida, obtenemos el 'Valor Mercado' y 'Pos_Especifica' correspondientes
    if resultado:
        nombre_coincidente = resultado[0]
        valor_mercado = df_transfermarkt[df_transfermarkt['Jugador_Normalizado'] == nombre_coincidente]['Valor Mercado'].values[0]
        pos_especifica = df_transfermarkt[df_transfermarkt['Jugador_Normalizado'] == nombre_coincidente]['Pos'].values[0]
        return valor_mercado, pos_especifica
    return None, None

# 5. Aplicar RapidFuzz solo a las filas que aún no tienen 'Valor Mercado' o 'Pos_Especifica'
df_combinado[['Valor Mercado', 'Pos_Especifica_Transfermarkt']] = df_combinado.apply(
    lambda row: pd.Series(mapear_por_nombre(row, df_transfermarkt)) 
    if pd.isna(row['Valor Mercado']) or pd.isna(row['Pos_Especifica_Transfermarkt']) else pd.Series([row['Valor Mercado'], row['Pos_Especifica_Transfermarkt']]),
    axis=1
)

# 6. Verificar las no coincidencias después de aplicar ambos pasos
no_coincidencias_finales = df_combinado['Valor Mercado'].isna().sum()
print(f"No coincidencias después de aplicar ambos pasos: {no_coincidencias_finales}")

# 7. Imprimir el número de coincidencias añadidas
coincidencias_antes = no_coincidencias - no_coincidencias_finales
print(f"Cantidad de coincidencias añadidas: {coincidencias_antes}")

# 8. Guardar el archivo final con todas las coincidencias
df_combinado.to_excel('/Users/macmontxinho/Desktop/Master Python/TFM/Union_Valores.xlsx', index=False)

# Mostrar las primeras filas para revisar el resultado
print(df_combinado.head())
```
