```python
def Fbref_disparos(url):

  headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'}
  page= (url)
  res = requests.get(page,headers=headers)
  comm = re.compile("<!--|-->")
  soup = BeautifulSoup(comm.sub("",res.text),'lxml')
  all_tables = soup.findAll("tbody")
  team_table = all_tables[0]
  a = all_tables[2]

  shots = a.find_all("td",{"data-stat": "shots"})
  shots_on_target = a.find_all("td",{"data-stat":"shots_on_target"})
  shots_on_target_pct = a.find_all("td",{"data-stat":"shots_on_target_pct"})
  shots_per90 = a.find_all("td",{"data-stat":"shots_per90"})
  shots_on_target_per90 = a.find_all("td",{"data-stat":"shots_on_target_per90"})
  goals_per_shot = a.find_all("td",{"data-stat":"goals_per_shot"})
  goals_per_shot_on_target = a.find_all("td",{"data-stat":"goals_per_shot_on_target"})
  average_shot_distance = a.find_all("td",{"data-stat":"average_shot_distance"})
  npxg_per_shot = a.find_all("td",{"data-stat":"npxg_per_shot"})

  ShotsList=[]
  Shots_on_targetList=[]
  Shots_on_target_pctList=[]
  Shots_per90List= []
  Shots_on_target_per90List=[]
  Goals_per_shotList=[]
  Goals_per_shot_on_targetList=[]
  Average_shot_distanceList=[]
  Npxg_per_shotList=[]

  for i in range(0, len(shots)):
      ShotsList.append(str(shots[i].get_text()))
  for i in range(0, len(shots_on_target)):
      Shots_on_targetList.append(str(shots_on_target[i].get_text()))
  for i in range(0, len(shots_on_target_pct)):
      Shots_on_target_pctList.append(str(shots_on_target_pct[i].get_text()))
  for i in range(0, len(shots_per90)):
      Shots_per90List.append(str(shots_per90[i].get_text()))
  for i in range(0, len(shots_on_target_per90)):
      Shots_on_target_per90List.append(str(shots_on_target_per90[i].get_text()))
  for i in range(0, len(goals_per_shot)):
      Goals_per_shotList.append(str(goals_per_shot[i].get_text()))
  for i in range(0, len(goals_per_shot_on_target)):
      Goals_per_shot_on_targetList.append(str(goals_per_shot_on_target[i].get_text()))
  for i in range(0, len(average_shot_distance)):
      Average_shot_distanceList.append(str(average_shot_distance[i].get_text()))
  for i in range(0, len(npxg_per_shot)):
      Npxg_per_shotList.append(str(npxg_per_shot[i].get_text()))

  df=pd.DataFrame({"Disparos":ShotsList,"Disparos_arco":Shots_on_targetList,"%Disparos_arco":Shots_on_target_pctList,"Disparos/90":Shots_per90List,
                              "Disparos_arco/90":Shots_on_target_per90List,"Gls/Disparos":Goals_per_shotList,"Gls/Disparos_arco":Goals_per_shot_on_targetList,"Distancia_disparo":Average_shot_distanceList})
  cols_num=['Disparos', 'Disparos_arco', '%Disparos_arco', 'Disparos/90',
       'Disparos_arco/90', 'Gls/Disparos', 'Gls/Disparos_arco',
       'Distancia_disparo']
  for col in cols_num:
      df[col]= pd.to_numeric(df[col],errors="coerce")

  df.fillna(0,inplace=True)

  return df
```

```python
def Fbref_tipos_pases(url):

  headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'}
  page= (url)
  res = requests.get(page,headers=headers)
  comm = re.compile("<!--|-->")
  soup = BeautifulSoup(comm.sub("",res.text),'lxml')
  all_tables = soup.findAll("tbody")
  team_table = all_tables[0]
  a = all_tables[2]

  passes_live = a.find_all("td",{"data-stat": "passes_live"})
  passes_dead = a.find_all("td",{"data-stat":"passes_dead"})
  passes_free_kicks = a.find_all("td",{"data-stat":"passes_free_kicks"})
  through_balls = a.find_all("td",{"data-stat":"through_balls"})
  passes_switches = a.find_all("td",{"data-stat":"passes_switches"})
  crosses = a.find_all("td",{"data-stat":"crosses"})
  throw_ins = a.find_all("td",{"data-stat":"throw_ins"})
  corner_kicks = a.find_all("td",{"data-stat":"corner_kicks"})
  corner_kicks_in = a.find_all("td",{"data-stat":"corner_kicks_in"})
  corner_kicks_out = a.find_all("td",{"data-stat":"corner_kicks_out"})
  corner_kicks_straight = a.find_all("td",{"data-stat":"corner_kicks_straight"})

  passes_offsides = a.find_all("td",{"data-stat":"passes_offsides"})
  passes_blocked = a.find_all("td",{"data-stat":"passes_blocked"})

  Passes_liveList=[]
  Passes_deadList=[]
  Passes_free_kicksList=[]
  Through_ballsList=[]
  Passes_switchesList=[]
  CrossesList=[]
  Throw_insList=[]
  Corner_kicksList=[]
  Corner_kicks_inList=[]
  Corner_kicks_outList=[]
  Corner_kicks_straightList=[]

  Passes_offsidesList=[]
  Passes_blockedList=[]

  for i in range(0, len(passes_live)):
    Passes_liveList.append(str(passes_live[i].get_text()))
  for i in range(0, len(passes_dead)):
    Passes_deadList.append(str(passes_dead[i].get_text()))
  for i in range(0, len(passes_free_kicks)):
    Passes_free_kicksList.append(str(passes_free_kicks[i].get_text()))
  for i in range(0, len(through_balls)):
    Through_ballsList.append(str(through_balls[i].get_text()))
  for i in range(0, len(passes_switches)):
    Passes_switchesList.append(str(passes_switches[i].get_text()))
  for i in range(0, len(crosses)):
    CrossesList.append(str(crosses[i].get_text()))
  for i in range(0, len(throw_ins)):
    Throw_insList.append(str(throw_ins[i].get_text()))
  for i in range(0, len(corner_kicks)):
    Corner_kicksList.append(str(corner_kicks[i].get_text()))
  for i in range(0, len(corner_kicks_in)):
    Corner_kicks_inList.append(str(corner_kicks_in[i].get_text()))
  for i in range(0, len(corner_kicks_out)):
    Corner_kicks_outList.append(str(corner_kicks_out[i].get_text()))

  for i in range(0, len(corner_kicks_straight)):
    Corner_kicks_straightList.append(str(corner_kicks_straight[i].get_text()))
  for i in range(0, len(passes_offsides)):
    Passes_offsidesList.append(str(passes_offsides[i].get_text()))
  for i in range(0, len(passes_blocked)):
    Passes_blockedList.append(str(passes_blocked[i].get_text()))

  df=pd.DataFrame({"Pase_balon_juego":Passes_liveList,"Pase_balon_allparado":Passes_deadList,"Pase_tiro_libre":Passes_free_kicksList,"Pase_tiro_libre":Passes_free_kicksList,"Pase_largo":Through_ballsList,"Cambios_juego":Passes_switchesList,
                            "Pase_cruzado":CrossesList,"Lanzamientos":Throw_insList,"Saques_esquina":Corner_kicksList,"Saque_esquina_dentro":Corner_kicks_inList,"Saque_esquina_fuera":Corner_kicks_outList,
                            "Saquee_esquina_recto":Corner_kicks_straightList,"Pase_fuera_juego":Passes_offsidesList,"Pase_bloqueado/fallado":Passes_blockedList})


  df.fillna(0,inplace=True)

  return df
```

```python
#Función para sacar cada liga

def fbref_liga(url,numberliga,liga):


  url_stat=url +str(numberliga)+ "/stats/" +str(liga) + "-Stats"
  url_disparos=url +str(numberliga)+ "/shooting/" +str(liga) + "-Stats"
  url_pases=url +str(numberliga)+ "/passing/" +str(liga) + "-Stats"
  url_tipos_pases=url +str(numberliga)+ "/passing_types/" +str(liga) + "-Stats"
  url_creacion_goles=url +str(numberliga)+ "/gca/" +str(liga) + "-Stats"
  url_acciones_defensivas=url +str(numberliga)+ "/defense/" +str(liga) + "-Stats"
  url_posesion=url +str(numberliga)+ "/possession/" +str(liga) + "-Stats"
  url_estadisticas_diversas=url +str(numberliga)+ "/misc/" +str(liga) + "-Stats"


  stat=fbref_stat(url_stat)
  disparos=Fbref_disparos(url_disparos)
  pases=Fbref_pases(url_pases)
  tipos_pases=Fbref_tipos_pases(url_tipos_pases)
  creacion_goles=Fbref_creacion_goles(url_creacion_goles)
  acciones_defensivas=Fbref_acciones_defensivas(url_acciones_defensivas)
  posesion=Fbref_posesion(url_posesion)
  estadisticas_diversas=Fbref_estadisticas_diversas(url_estadisticas_diversas)


  df=pd.concat([stat,disparos,pases,tipos_pases,creacion_goles,acciones_defensivas,posesion,estadisticas_diversas],axis=1)


  return df
```

```python
def limpiar_BBDD(df_jugs):
    df_jugs['Nacionalidad'] = df_jugs['Nacionalidad'].str.split().str[-1]

    cols_espacios = df_jugs.columns[df_jugs.applymap(lambda x: x == '').any()]
    df_jugs[cols_espacios[1:]] = df_jugs[cols_espacios[1:]].replace('', 0)

    df_jugs['Minutos'] = df_jugs['Minutos'].astype(str).str.replace(',', '').astype(int)


    cols_int = ['Partidos', 'P.inicio', 'Pases_exito', 'Pases_intentados',  'Distancia_total_pase', 'Distancia_total_progresiva', 'Pase_corto_exito',
                'Pase_corto_intentado', 'Pase_medio_exito', 'Pase_medio_intentado', 'Pase_largo_exito', 'Pase_largo_intentado', 'Pases_clave',
                'Pases_exito_1/3', 'Pases_exito_area', 'Centros_area_exito', 'Pases_progresivos', 'Pase_balon_juego', 'Pase_balon_allparado',
                'Pase_tiro_libre', 'Pase_largo', 'Cambios_juego', 'Pase_cruzado', 'Lanzamientos', 'Saques_esquina', 'Saque_esquina_dentro',
                'Saque_esquina_fuera', 'Saquee_esquina_recto', 'Pase_fuera_juego', 'Pase_bloqueado/fallado', 'SCA', 'SCA_pase_on', 'SCA_pase_off',
                'SCA_toma_exitosa', 'SCA_disparo', 'SCA_tiro_libre', 'SCA_defensa', 'ACG', 'ACG_pase_on', 'ACG_pase_off', 'ACG_acciones', 'ACG_disparos',
                'ACG_tiro_libre', 'ACG_defensa', 'Tkl_intentados', 'Tkl_ganados', 'Tkl_1/3', 'Tkl_2/3', 'Tkl_3/3', 'Desafios_ganados', 'Desafios_intentados',
                'Desafios_perdidos', 'Bloqueos', 'Bloqueros_disparo', 'Bloqueo_pases', 'Intercepciones', 'Tkl_Intercepciones', 'Despejes', 'Errores',
                'Toques_balón', 'Toques_area_penal_def', 'Toques_1/3', 'Toques_2/3', 'Toques_3/3', 'Toques_area_penal_att', 'Toques_balon_juego',
                'Regates_intentados', 'Regates_exito', 'Tkld', 'Transportes_balon', 'Distancia_transportes_balon', 'Distanca_trasporte_progresivo',
                'Acarreos_progresivos', 'Transportes_3/3', 'Tralados_area_penal_att', 'Errores_control_no_forzado', 'Error_control_forzado',
                'Pases_recibidos', 'Pases_progresivos_recibidos', 'TA', 'TR']
    df_jugs[cols_int] = df_jugs[cols_int].astype(int)

    df_jugs[['Edad', 'Nacimiento']] = df_jugs[['Edad', 'Nacimiento']].astype(int)

    cols_int2 = ['2TA', 'Fls_cometidas', 'Fls_recibidas', 'Fueras_de_juego', 'Gls_contra', 'Balon_recuperado', 'Arereo_ganado', 'Aereo_perdido']
    df_jugs[cols_int2] = df_jugs[cols_int2].astype(int)

    cols_pct = df.filter(like='%').columns
    df_jugs[cols_pct] = np.round(df_jugs[cols_pct] / 100, 2)

    df_jugs = df_jugs.reset_index(drop=True)
    return df_jugs
```
